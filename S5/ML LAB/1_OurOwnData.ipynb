```python
import pandas as pd
file=pd.read_csv("/content/13 MLP winequality-red.csv")
x=file[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides"]]
y=file["quality"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
from sklearn.linear_model import Perceptron
name=Perceptron()
name.fit(xtrain,ytrain)
op=name.predict(xtest)
from sklearn.metrics import accuracy_score
ac=accuracy_score(ytest,op)
print(ac)
op2=name.predict([[9.2,0.68,0.006,1.9,0.061]])
print(op2)
```

```python
import pandas as pd
file=pd.read_csv("/content/irisexcel.csv")
x=file[["sepal_length","sepal_width","petal_length","petal_width"]]
y=file["species"]
from sklearn.linear_model import Perceptron
josh=Perceptron()
josh.fit(x,y)
out=josh.predict([[1.2,2.3,3.1,3.3]])
print (out)

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(ytest,op)
print(cm)

cmd=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=josh.classes_)
cmd.plot()
```
